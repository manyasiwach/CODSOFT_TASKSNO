# Task 3: Data Visualization

## Objective

The objective of this task is to create meaningful visualizations using the Iris dataset. Different types of charts are used to understand the distribution of data, compare features, and identify relationships between variables.

```python
# Import Required Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_iris
```

```python
# Load Iris Dataset

iris = load_iris()

df = pd.DataFrame(
    iris.data,
    columns=iris.feature_names
)

df["Species"] = iris.target

df["Species"] = df["Species"].map({
    0: "Setosa",
    1: "Versicolor",
    2: "Virginica"
})

df.head()
```

```python
plt.figure(figsize=(6,4))

sns.countplot(data=df, x="Species")

plt.title("Number of Flowers in Each Species")
plt.xlabel("Species")
plt.ylabel("Count")

plt.show()
```

```python
plt.figure(figsize=(6,4))

plt.hist(df["petal length (cm)"], bins=10)

plt.title("Distribution of Petal Length")
plt.xlabel("Petal Length")
plt.ylabel("Frequency")

plt.show()
```

```python
# Average Sepal Length of Each Species

avg_sepal = df.groupby("Species")["sepal length (cm)"].mean()

plt.figure(figsize=(6,4))

plt.plot(avg_sepal.index, avg_sepal.values, marker='o')

plt.title("Average Sepal Length by Species")
plt.xlabel("Species")
plt.ylabel("Average Sepal Length")

plt.show()
```

```python
# Species Distribution

species_count = df["Species"].value_counts()

plt.figure(figsize=(6,6))

plt.pie(
    species_count,
    labels=species_count.index,
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Species Distribution")

plt.show()
```

```python
# Scatter Plot

plt.figure(figsize=(7,5))

sns.scatterplot(
    data=df,
    x="sepal length (cm)",
    y="petal length (cm)",
    hue="Species"
)

plt.title("Sepal Length vs Petal Length")

plt.show()
```

```python
# Correlation Heatmap

plt.figure(figsize=(8,6))

sns.heatmap(
    df.iloc[:,0:4].corr(),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Between Features")

plt.show()
```

# Insights

- The dataset contains three balanced Iris species.
- Petal length and petal width are strongly positively correlated.
- Setosa is clearly distinguishable from the other two species.
- Visualizations help identify patterns and relationships within the dataset.
- These charts make the dataset easier to understand for analysis and decision-making.
