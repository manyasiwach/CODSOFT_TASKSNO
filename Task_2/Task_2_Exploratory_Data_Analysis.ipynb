# Task 2: Exploratory Data Analysis (EDA)

## Objective

The objective of this task is to perform Exploratory Data Analysis (EDA) on the Iris dataset by examining descriptive statistics, identifying trends, understanding relationships between variables, detecting outliers, and drawing meaningful insights using Python.

```python
# Import Required Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_iris
```

```python
# Load Iris Dataset

iris = load_iris()

df = pd.DataFrame(
    iris.data,
    columns=iris.feature_names
)

df["Species"] = iris.target

df["Species"] = df["Species"].map({
    0: "Setosa",
    1: "Versicolor",
    2: "Virginica"
})

df.head()
```

```python
# Dataset Overview

print("Shape of Dataset:", df.shape)

print("\nColumn Names:")
print(df.columns)

print("\nDataset Information:")
df.info()
```

```python
# Descriptive Statistics

df.describe()
```

```python
# Distribution of Species

df["Species"].value_counts()
```

```python
# Histogram of all numerical features

df.hist(figsize=(10,8))
plt.suptitle("Distribution of Iris Dataset Features")
plt.show()
```

```python
# Box Plot to Detect Outliers

plt.figure(figsize=(10,6))

sns.boxplot(data=df.iloc[:,0:4])

plt.title("Box Plot for Outlier Detection")

plt.show()
```

```python
# Correlation Heatmap

plt.figure(figsize=(8,6))

sns.heatmap(df.iloc[:,0:4].corr(),
            annot=True,
            cmap="Blues")

plt.title("Correlation Heatmap")

plt.show()
```

```python
# Relationship Between Sepal Length and Petal Length

plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df,
    x="sepal length (cm)",
    y="petal length (cm)",
    hue="Species"
)

plt.title("Sepal Length vs Petal Length")

plt.show()
```

```python
# Pair Plot

sns.pairplot(df, hue="Species")

plt.show()
```

# Observations

- The dataset contains three Iris species: Setosa, Versicolor, and Virginica.
- No missing values are present.
- The dataset is balanced, with an equal number of samples for each species.
- Petal length and petal width show a strong positive correlation.
- Setosa is clearly separated from the other two species.
- A few potential outliers can be observed in some numerical features.

```python
# Summary Statistics

print(df.describe())
```
