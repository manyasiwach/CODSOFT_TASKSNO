# Task 1: Data Cleaning and Preprocessing

## Objective
The objective of this task is to clean and preprocess the Iris dataset by inspecting its structure, checking missing values, removing duplicate records, verifying data types, and preparing a clean dataset for further analysis.

```python
# Task 1 : Data Cleaning & Preprocessing

import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
```

```python
# Load Iris Dataset

iris = load_iris()

df = pd.DataFrame(
    iris.data,
    columns=iris.feature_names
)

df["Species"] = iris.target

df["Species"] = df["Species"].map({
    0: "Setosa",
    1: "Versicolor",
    2: "Virginica"
})

df.head()
```

```python
# Inspect Dataset Structure

print("Shape of Dataset:", df.shape)

print("\nColumn Names:")
print(df.columns)

print("\nDataset Information:")
df.info()
```

```python
# Check Missing Values

print("Missing Values in Each Column:")
print(df.isnull().sum())
```

```python
# Check Duplicate Records

duplicates = df.duplicated().sum()

print("Total Duplicate Records:", duplicates)
```

```python
# Remove Duplicate Records

df = df.drop_duplicates()

print("New Shape of Dataset:", df.shape)
```

```python
# Check Data Types

print(df.dtypes)
```

```python
# Check Unique Values in Species Column

print(df["Species"].unique())
```

```python
# Save Cleaned Dataset

df.to_csv("iris_cleaned.csv", index=False)

print("Cleaned Dataset Saved Successfully!")
```

```python
# Final Summary

print("========== DATA CLEANING SUMMARY ==========")
print(f"Total Rows after Cleaning : {df.shape[0]}")
print(f"Total Columns             : {df.shape[1]}")
print("\nMissing Values:")
print(df.isnull().sum())
print("\nDuplicate Records:", df.duplicated().sum())
print("\nData Types:")
print(df.dtypes)
print("\nDataset is cleaned and ready for further analysis.")
```

# Conclusion

The Iris dataset was successfully imported and inspected.

The dataset structure, column names, and data types were examined.

Missing values and duplicate records were checked.

Duplicate records were removed to improve data quality.

The cleaned dataset was saved as **iris_cleaned.csv**.

The dataset is now ready for further analysis and visualization.
